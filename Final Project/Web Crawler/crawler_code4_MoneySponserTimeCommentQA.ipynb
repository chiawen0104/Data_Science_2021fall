```python
pj_sponsor_num_ended = []
pj_target = []
pj_time_start = []
pj_time_end = []
pj_comment_num = []
pj_QA_num = []

for i in range(len(pj_url_ended)):
  res = requests.get(pj_url_ended[i], headers = headers)
  soup = BeautifulSoup(res.text, 'html.parser')

  # 贊助人數
  sponsor_num = soup.find(class_= 'mb1 f7').find(class_ = 'js-backers-count').text

  # 目標金額(與群眾集資方法不同)
  temp = soup.find(class_="f3 b js-sum-raised nowrap").find_next_siblings('div')
  for div in temp:
    target = div.text[6:].strip()

  # 時程 開始:結束
  time = soup.find(class_ = 'mb2 f7').text
  time_start = time[3:20].strip()
  time_end = time[23:].strip()

  # 留言數&問答數
  text = soup.find(class_ = 'mh3 mh0-ns ph0 ph3-ns w-70-l overflow-auto nowrap flex flex-nowrap').text.strip()
  a = text.find('言')
  b = text.find('常')
  c = text.find('答')
  d = text.find('馬')
  if b == -1: # 沒有常見問答
    comment_num = text[a+1:].strip()
    QA_num = 'N/A'
  elif d == -1: # 沒有馬上買
    comment_num = text[a+1:b].strip()
    QA_num = text[c+1:].strip()
  else: # 有馬上買
    comment_num = text[a+1:b].strip()
    QA_num = text[c+1:d].strip() 
  
  pj_sponsor_num_ended.append(sponsor_num)
  pj_target.append(target)
  pj_time_start.append(time_start)
  pj_time_end.append(time_end)
  pj_comment_num.append(comment_num)
  pj_QA_num.append(QA_num)


for i in range(0, num_pj_ended):
  all_pj_ended[i]['sponsor_num'] = pj_sponsor_num_ended[i]
  all_pj_ended[i]['target'] = pj_target[i]
  all_pj_ended[i]['time_start'] = pj_time_start[i]
  all_pj_ended[i]['time_end'] = pj_time_end[i]
  all_pj_ended[i]['comment_num'] = pj_comment_num[i]
  all_pj_ended[i]['QA_num'] = pj_QA_num[i]

print(pj_sponsor_num_ended)
print(pj_target)
print(pj_time_start)
print(pj_time_end)
print(pj_comment_num)
print(pj_QA_num)

print(all_pj_ended[10])
print(all_pj_ended)
```
